# Time Series EDA 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as s
%matplotlib inline 
```

```python
df = pd.read_csv("Sales.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month',inplace=True)
```

```python
df.info()
```

```python
df.tail()
```

Identify max Sales each year 

```python
df['Sales'].resample(rule='YE').max().plot(figsize=(15,8),marker='o')
plt.title("Max Sales Per Month")
plt.show()
```

```python
df['Sales'].resample(rule='YE').max()
```

Identify Mean sales for Each Year 

```python
df['Sales'].resample(rule='YE').mean()
```

```python
df['Sales'].resample(rule='YE').mean().plot(figsize=(12,8),marker = 'o',c = 'Purple')
```

Identify Minimum sales for each year 

```python
df['Sales'].resample(rule='YE').min()
```

```python
df['Sales'].resample(rule='YE').min().plot(figsize=(12,8),marker='o',c = 'violet')
```

Sum Sales in Each 5-Year Period

```python
df['Sales'].resample(rule='5Y').sum()
```

```python
df['Sales'].resample(rule='5Y').sum().plot(figsize=(12,8),marker='o')
```
